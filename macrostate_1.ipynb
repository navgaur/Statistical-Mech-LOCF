# **Plot the probability of various macrostates in coin-tossing experiment**

This is a two level system as the coin tossing output is of two levels (heads or tails)

The plot has to be of Probability of macrostate as a function of (say) number of heads. The plot has to be for 4, 8, 16 coins etc.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Function to calculate binomial probability mass function
def binomial(n, k, p):
  return binom.pmf(k, n, p)

# Number of coin tosses
n_values = [4, 8, 16, 32]

# Probability of getting heads in a single toss
p = 0.5

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the probability distribution for each number of coin tosses
for n in n_values:
  k_values = np.arange(0, n + 1)
  probabilities = binomial(n, k_values, p)
  ax.plot(k_values, probabilities, label=f'n={n}')

# Set plot labels and title
ax.set_xlabel('Number of Heads')
ax.set_ylabel('Probability')
ax.set_title('Macrostate Probability Distribution of Number Heads in Coin Tossing')
ax.legend()

# Show the plot
plt.show()
```
