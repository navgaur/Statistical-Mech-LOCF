# **Plot the probability of various macrostates in coin-tossing experiment**

This is a two level system as the coin tossing output is of two levels (heads or tails)

The plot has to be of Probability of macrostate as a function of (say) number of heads. The plot has to be for 4, 8, 16 coins etc.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Function to calculate binomial probability mass function
def binomial(n, k, p):
  return binom.pmf(k, n, p)

# Number of coin tosses
n_values = [4, 8, 16, 32]

# Probability of getting heads in a single toss
p = 0.5

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the probability distribution for each number of coin tosses
for n in n_values:
  k_values = np.arange(0, n + 1)
  probabilities = binomial(n, k_values, p)
  ax.plot(k_values, probabilities, label=f'n={n}')

# Set plot labels and title
ax.set_xlabel('Number of Heads')
ax.set_ylabel('Probability')
ax.set_title('Macrostate Probability Distribution of Number Heads in Coin Tossing')
ax.legend()

# Show the plot
plt.show()
```

```python
import random
import itertools
import matplotlib.pyplot as plt

def coin_toss_experiment(num_coins, num_trials):
    macrostate_counts = {}

    outcomes = ['H', 'T']
    all_possible_outcomes = [''.join(p) for p in itertools.product(outcomes, repeat=num_coins)]

    for _ in range(num_trials):
        result = ''.join(random.choice(outcomes) for _ in range(num_coins))
        num_heads = result.count('H')
        num_tails = num_coins - num_heads
        macrostate = f'{num_heads}H-{num_tails}T'
        macrostate_counts[macrostate] = macrostate_counts.get(macrostate, 0) + 1

    total_trials = float(num_trials)
    probabilities = {state: count / total_trials for state, count in macrostate_counts.items()}

    # Sort macrostates for better visualization
    sorted_states = sorted(probabilities.keys(), key=lambda x: (int(x.split('H-')[0].rstrip('H')), int(x.split('-')[1].rstrip('T'))))

    sorted_probabilities = {state: probabilities[state] for state in sorted_states}

    return sorted_probabilities

def plot_probabilities(probabilities):
    states = list(probabilities.keys())
    probs = list(probabilities.values())

    plt.bar(states, probs)
    plt.xlabel('Macrostates')
    plt.ylabel('Probability')
    plt.title('Probability of Macrostates in Coin-Tossing Experiment')
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Set the number of coins and trials
num_coins = 8
num_trials = 1000

# Run the coin-toss experiment
probabilities = coin_toss_experiment(num_coins, num_trials)

# Plot the probabilities
plot_probabilities(probabilities)
```

```python
import random
import matplotlib.pyplot as plt

outcomes=['H','T']
no_coins=4

no_of_events=1000

list_heads=[]
for _ in range(no_of_events):
  t1=[]
  for _ in range(no_coins):
    t1.append(random.choice(outcomes))
  outcome=''.join(t1)
  no_heads=outcome.count('H')
  list_heads.append(no_heads)

macro_state=[0,1,2,3,4]

#initialize dictionary
for i in macro_state:
  macro_dict[i] = 0

for i in list_heads:
  if i in macro_state:
    macro_dict[i] += 1

print(macro_dict)

keys = macro_dict.keys()
values = macro_dict.values()
values = [t/no_of_events for t in values]

plt.plot(keys,values,marker='o',linestyle="--")
plt.show()
```
