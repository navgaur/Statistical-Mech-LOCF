#**Plot Planck’s law of Black body radiation w.r.t. wavelength/frequency at different temperatures. Compare it with Rayleigh-Jeans Law and Wien’s distribution law for a given temperature**

##**Planck's law of black body radiaion**

Energy density associated with frequency range $(\omega, \omega + d\omega)$

(**Statistical Mechanics, R. K. Pathria, 3rd Ed, Page 200**)

$$u(\nu,T) d\nu= \frac{8 \pi h \nu^3}{c^3}   \frac{d \nu}{exp\left(\frac{h \nu}{k T}\right) - 1} $$


##**Rayleigh-Jeans law**

for longer wavelengths or smaller frequencies ($h\nu \ll kT$) we have
$$\frac{h\nu}{kT} \ll 1$$
$$e^{h\nu}{kT} \approx 1 + \frac{h\nu}{kT}$$

Energy density:

$$u(\nu,T) = \frac{8 \pi h \nu^3}{c^3}   \frac{1}{\frac{h \nu}{kT}}
= \frac{8 \pi \nu^2 kT}{c^3} $$


##**Wien's distribution law**

for shorter wavelenghts or larger frequencies ($h\nu \gg kT$) we have
$$exp\left(\frac{h\nu}{kT}\right) - 1 \approx exp\left(\frac{h\nu}{kT}\right) $$

$$u(\nu,T) = \frac{8 \pi h \nu^3}{c^3} exp\left(- \frac{h \nu}{k T}\right) $$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import Boltzmann, Planck, speed_of_light, pi

k=Boltzmann
h=Planck
c=speed_of_light

# Frequency range
f = np.linspace(1e13, 5e14, 1000)

# Planck's Law for energy density
def u_planck(f,T):
  return (8 * pi * h * f**3) / (c**3 * (np.exp((h * f) / (k * T)) - 1))

def u_rayleigh(f,T):
  return (8 * pi *k * T * f**2) / (c**3)

def u_wiens(f,T):
  return (8 * pi * h * f**3) * np.exp(-(h * f) / (k * T)) / (c**3)

# Plotting
plt.figure(figsize=(10, 6))
temperature=[1000,1500,2000]

for T in temperature:
  print(T)
  t1=u_planck(f,T)
  plt.plot(f, t1, label=f'Temperature = {T} K')

plt.title('Planck\'s Law - Energy Density vs. Frequency')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Energy Density (J/m³/Hz)')
#plt.xscale("log")
#plt.yscale("log")
plt.legend()
plt.grid(True)
plt.show()
```
