#**Plot Specific Heat of Solids w.r.t temperature**

 *Classical value* of specific heat $C_V = 3 R$ with $R$ is gas constant. This is also known as **Dulong-Petit Law**

##**Debye Distribution function**
(Pathria, Page-209)

$$C_V(T) = 3 N k D(x_0)$$
 - N is number of atoms per mole (Avogadro's Number)
 - $k$ is Boltzmann constant
 - $\Theta_D$ is Debye Temperature

with $D(x_0)$ called *Debye Function*
$$
D(x_0)= \frac{3}{x_0^3} \int_0^{x_0} \frac{x^4 e^x}{\left(e^x - 1\right)^2} dx
$$
$$x_0 = \frac{\Theta_D}{T}$$
where $\Theta_D$ is called *Debye Temperature* of the solid


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import Boltzmann
from scipy.integrate import quad

k = Boltzmann
N = 6.022e23

# Define temperature range
temp = np.linspace(1, 500, 100)

# Debye temperature (in Kelvin)
Debye_temp = 300.0

# Debye function integrand
def DD(x):
    return x**4 * np.exp(x) / (np.exp(x) - 1)**2

const = 3
res1 = []
for T in temp:
  x0 = Debye_temp/T
  t1 = const* (1/x0**3) * quad(DD,0,x0)[0]
  res1.append(t1)

# Calculate the Debye model for the given temperature range
#debye_function = 9 * len(temp) * 8.314 * (temp / Debye_temp)**3 * res1

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(temp, res1, label=f'Debye Temperature = {Debye_temp} K', color='green')
plt.xlabel('Temperature (K)')
plt.ylabel('$C_V/3Nk$')
plt.title('Heat Capacity of Solids')
plt.legend()
plt.grid(True)
plt.show()
```

##**Einstein Distribution function**
 (Pathria, Page-209)

 $$ C_V(T) = 3 N k E(x) $$

 With $E(x)$ is called *Einstein function*
 $$ E(x) = \frac{x^2 e^x}{(e^x - 1)^2}$$
 with $x = \Theta_E/T$ and $\Theta_E$ is *Einstein Temperature*
