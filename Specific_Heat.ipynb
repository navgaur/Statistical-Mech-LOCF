```python
import numpy as np
import matplotlib.pyplot as plt

# Define temperature range
temperature_range = np.linspace(0.1, 10, 100)  # Example range from 0.1 K to 10 K

# Define the Einstein temperature (in Kelvin)
Einstein_temperature = 1.0  # Example value

# Calculate the Einstein function for the given temperature range
einstein_function = 3 * (Einstein_temperature / temperature_range)**2 * np.exp(Einstein_temperature / temperature_range) / (np.exp(Einstein_temperature / temperature_range) - 1)**2

# Dulong Petit
def dulong(temp):
  return 3 * 8.314

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(temperature_range,dulong(temperature_range))
plt.plot(temperature_range, einstein_function, label=f'Einstein Temperature = {Einstein_temperature} K', color='blue')
plt.xlabel('Temperature (K)')
plt.ylabel('Einstein Function')
plt.title('Einstein Function for Heat Capacity of Solids')
plt.legend()
plt.grid(True)
plt.show()
```
